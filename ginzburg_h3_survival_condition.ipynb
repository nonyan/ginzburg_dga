```python
def generate_quiver_lists(max_c):
    """
    1 <= a <= b <= c <= max_c の範囲で、
    コホモロジーが生き残る条件 (abc <= b^2 + c^2 - 2) を満たすリストと
    満たさないリストを分けて生成する関数。
    """
    surviving_list = [] # abc <= b^2 + c^2 - 2 を満たす (ジャンプする/生き残る)
    stable_list = []    # abc >  b^2 + c^2 - 2 を満たす (安定する/消える)
    
    for a in range(1, max_c + 1):
        for b in range(a, max_c + 1):
            for c in range(b, max_c + 1):
                # 条件の判定
                if a * b * c <= b**2 + c**2 - 2:
                    surviving_list.append((a, b, c))
                else:
                    stable_list.append((a, b, c))
                    
    return surviving_list, stable_list

# ===== 実行例 =====
bound = 5
survive, stable = generate_quiver_lists(bound)

print(f"=== cの上限が {bound} の場合 ===")
print(f"【生き残るリスト (abc <= b^2 + c^2 - 2)】 計 {len(survive)} 個")
# aの値ごとに見やすく表示
for current_a in range(1, bound + 1):
    items_for_a = [item for item in survive if item[0] == current_a]
    if items_for_a:
        print(f"  a={current_a}: {items_for_a}")

print(f"\n【安定するリスト (abc > b^2 + c^2 - 2)】 計 {len(stable)} 個")
for current_a in range(1, bound + 1):
    items_for_a = [item for item in stable if item[0] == current_a]
    if items_for_a:
        print(f"  a={current_a}: {items_for_a}")
```
