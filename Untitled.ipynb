```python
import numpy as np
import itertools

def solve_ginzburg_naive(a, b, c, max_degree=5):
    """
    NumPyを用いた行列ランク計算によるJacobian代数の次元計算
    (SageMath不要、Python環境だけで動作可能)
    """
    print(f"--- Analyzing Q({a}, {b}, {c}) with Linear Algebra ---")
    
    # 1. 矢印の定義 (名前と始点・終点)
    # 1->2 (x), 2->3 (y), 3->1 (z)
    arrows = []
    for i in range(a): arrows.append({'name': f'x{i+1}', 's': 1, 'e': 2, 'type': 'x', 'idx': i})
    for i in range(b): arrows.append({'name': f'y{i+1}', 's': 2, 'e': 3, 'type': 'y', 'idx': i})
    for i in range(c): arrows.append({'name': f'z{i+1}', 's': 3, 'e': 1, 'type': 'z', 'idx': i})
    
    arrow_map = {ar['name']: i for i, ar in enumerate(arrows)}
    num_arrows = len(arrows)

    # ランダムな係数 (GenericなWを再現)
    # W = sum C_ijk * xi * yj * zk
    # 固定シードで再現性確保
    np.random.seed(42)
    # 係数は複素数だと重いので、有限体風にランダムな整数を使う（ランク計算には十分）
    Coeffs = np.random.randint(1, 100, size=(a, b, c))

    # --- 基本的な関係式 (Jacobian relations: degree 2) の生成 ---
    # format: [(係数, [矢印インデックスのリスト]), ...]
    base_relations = []

    # dW/dx_i = sum C_ijk * y_j * z_k = 0
    for i in range(a):
        rel = []
        for j in range(b):
            for k in range(c):
                # y_j * z_k というパス
                y_idx = a + j # arrowsリスト内でのyのインデックス
                z_idx = a + b + k
                coeff = Coeffs[i, j, k]
                rel.append((coeff, [y_idx, z_idx]))
        base_relations.append({'s': 2, 'e': 1, 'terms': rel}) # 始点2, 終点1のパス

    # dW/dy_j = sum C_ijk * z_k * x_i = 0
    for j in range(b):
        rel = []
        for i in range(a):
            for k in range(c):
                z_idx = a + b + k
                x_idx = i
                coeff = Coeffs[i, j, k]
                rel.append((coeff, [z_idx, x_idx]))
        base_relations.append({'s': 3, 'e': 2, 'terms': rel})

    # dW/dz_k = sum C_ijk * x_i * y_j = 0
    for k in range(c):
        rel = []
        for i in range(a):
            for j in range(b):
                x_idx = i
                y_idx = a + j
                coeff = Coeffs[i, j, k]
                rel.append((coeff, [x_idx, y_idx]))
        base_relations.append({'s': 1, 'e': 3, 'terms': rel})

    # --- 次数ごとの計算 ---
    
    # 次数1は単に矢印の数
    print(f"Degree 1: dim = {len(arrows)}")

    for d in range(2, max_degree + 1):
        # 1. 長さ d の有効なパスをすべて列挙
        valid_paths = [] # list of tuple of indices
        
        # 再帰的にパスを生成
        def generate_paths(current_path, current_end_node):
            if len(current_path) == d:
                valid_paths.append(tuple(current_path))
                return
            
            # 次につながる矢印を探す
            for i, ar in enumerate(arrows):
                if ar['s'] == current_end_node:
                    generate_paths(current_path + [i], ar['e'])

        # 始点は1,2,3どれでもよい
        for i, ar in enumerate(arrows):
            generate_paths([i], ar['e'])
            
        num_paths = len(valid_paths)
        if num_paths == 0:
            print(f"Degree {d}: dim = 0")
            break

        # パスIDのマッピング (行列の列に対応)
        path_to_col = {p: i for i, p in enumerate(valid_paths)}
        
        # 2. 関係式行列の構築
        # 行: 制約式, 列: パス
        # base_relations (長さ2) に、左から長さ L、右から長さ R のパスを掛けて、全長 d になるものを作る
        
        constraints = []
        
        # 長さ2の関係式なので、残り長さは d - 2
        remainder = d - 2
        
        # 左側に長さ left_len, 右側に長さ right_len のパスを付加 (left+right = remainder)
        for left_len in range(remainder + 1):
            right_len = remainder - left_len
            
            # 左側パスの生成
            left_paths = []
            if left_len == 0:
                left_paths = [([], n) for n in [1,2,3]] # (path, end_node) 始点は任意
            else:
                # 簡易的なパス生成（全探索）
                # 注意: 効率化のため本当はキャッシュすべきだが、d=5程度なら全探索でOK
                temp_paths = []
                def get_p(cp, ce):
                    if len(cp) == left_len: temp_paths.append((cp, ce)); return
                    for i, ar in enumerate(arrows):
                        if ar['s'] == ce: get_p(cp+[i], ar['e'])
                for n in [1,2,3]:
                    for i, ar in enumerate(arrows):
                        if ar['s'] == n: get_p([i], ar['e'])
                left_paths = temp_paths

            for l_p, l_end in left_paths:
                # この左側パスの終点 l_end から始まる relation を探す
                for rel in base_relations:
                    if rel['s'] == l_end:
                        # 右側パスの生成 (relationの終点 rel['e'] から始まる)
                        right_paths = []
                        if right_len == 0:
                            right_paths = [[]]
                        else:
                            temp_r = []
                            def get_r(cp, ce):
                                if len(cp) == right_len: temp_r.append(cp); return
                                for i, ar in enumerate(arrows):
                                    if ar['s'] == ce: get_r(cp+[i], ar['e'])
                            # relationの終点からスタート
                            for i, ar in enumerate(arrows):
                                if ar['s'] == rel['e']: get_r([i], ar['e'])
                            right_paths = temp_r
                        
                        # 制約式を生成 (左パス + relation項 + 右パス)
                        for r_p in right_paths:
                            # 1つの制約式（行ベクトル）を作る
                            row = np.zeros(num_paths)
                            for coeff, term_indices in rel['terms']:
                                # 全体のパス: l_p + term + r_p
                                full_path = tuple(l_p + term_indices + r_p)
                                if full_path in path_to_col:
                                    col_idx = path_to_col[full_path]
                                    row[col_idx] = coeff
                            constraints.append(row)

        if not constraints:
            print(f"Degree {d}: dim = {num_paths} (No constraints)")
            continue

        # 3. ランク計算
        matrix = np.array(constraints)
        # 行列が大きいと遅いが、d=5程度なら大丈夫
        rank = np.linalg.matrix_rank(matrix)
        
        dim = num_paths - rank
        print(f"Degree {d}: dim = {dim} (Paths: {num_paths}, Rank: {rank})")
        
        if dim == 0:
            print("  -> Algebra is finite dimensional.")
            break

# --- 実行 ---
solve_ginzburg_naive(1, 1, 1, max_degree=6)
print("-" * 20)
solve_ginzburg_naive(1, 2, 3, max_degree=5)
print("-" * 20)
solve_ginzburg_naive(2, 2, 2, max_degree=4)
```

```python
import numpy as np
import itertools

def check_algebra_growth_sweep(a_max=3, b_max=3, c_max=3, max_check_degree=5):
    """
    a,b,c を指定範囲で動かし、Jacobian代数が有限次元か無限次元かを判定する。
    """
    print(f"--- Sweeping parameters a,b,c up to ({a_max},{b_max},{c_max}) ---")
    print(f"{'Triple (a,b,c)':<15} | {'Markov?':<8} | {'Status':<12} | {'Dim Sequence (deg 1->'+str(max_check_degree)+')'}")
    print("-" * 85)

    # 重複を避けるため、a <= b <= c の組み合わせのみを計算してもよいが、
    # 念のため全探索し、表示を見やすくする
    for a in range(1, a_max + 1):
        for b in range(1, b_max + 1):
            for c in range(1, c_max + 1):
                
                # 計算コスト削減: 対称性から a<=b<=c の場合のみ計算し、他は省略することも可能だが
                # 今回は高速なので全て回す
                
                # --- Markov Triple Check ---
                # a^2 + b^2 + c^2 = k * a * b * c となる整数 k があるか？
                lhs = a**2 + b**2 + c**2
                rhs_base = a * b * c
                is_markov = False
                markov_k = ""
                
                if rhs_base > 0 and lhs % rhs_base == 0:
                    k = lhs // rhs_base
                    is_markov = True
                    markov_k = f"(k={k})"
                
                # --- Dimension Calculation (Linear Algebra) ---
                dims = get_degree_dims(a, b, c, max_check_degree)
                
                # --- Result Classification ---
                if dims[-1] == 0:
                    status = "Finite"
                elif len(dims) >= 2 and dims[-1] > dims[-2]:
                    status = "Growing" # 無限かつ増大
                else:
                    status = "Const/Inf" # 無限だが増大しない（あるいは振動）

                markov_str = "Yes" + markov_k if is_markov else ""
                dims_str = str(dims)
                
                print(f"({a}, {b}, {c})".ljust(15) + f" | {markov_str:<8} | {status:<12} | {dims_str}")

def get_degree_dims(a, b, c, max_degree):
    """
    特定の(a,b,c)について各次数の次元リストを返す
    """
    # 1. 矢印定義
    arrows = []
    for i in range(a): arrows.append({'s': 1, 'e': 2})
    for i in range(b): arrows.append({'s': 2, 'e': 3})
    for i in range(c): arrows.append({'s': 3, 'e': 1})
    num_arrows = len(arrows)
    
    if num_arrows == 0: return [0] * max_degree

    # 係数ランダム生成
    np.random.seed(a + b*10 + c*100) # 再現性のためシード固定
    # 簡易的に0,1の係数ではなく、ランダムな整数係数を用いる
    # ランク計算において「偶然0になる」のを防ぐため
    Coeffs = np.random.randint(1, 50, size=(a, b, c))

    # 関係式 (Jacobian derivatives) の生成
    # format: {'s': start_node, 'e': end_node, 'terms': [(coeff, [idx_2nd_arrow, idx_3rd_arrow])]}
    # 注意: ここではパスの長さ2の関係式を作る
    base_relations = []
    
    # 矢印のインデックス範囲
    x_range = range(0, a)
    y_range = range(a, a+b)
    z_range = range(a+b, a+b+c)
    
    # dW/dx (start:2 -> end:1 via y, z)
    for i in range(a):
        terms = []
        for j in range(b):
            for k in range(c):
                terms.append((Coeffs[i, j, k], [a+j, a+b+k])) # y_j, z_k
        base_relations.append({'s': 2, 'e': 1, 'terms': terms})

    # dW/dy (start:3 -> end:2 via z, x)
    for j in range(b):
        terms = []
        for i in range(a):
            for k in range(c):
                terms.append((Coeffs[i, j, k], [a+b+k, i])) # z_k, x_i
        base_relations.append({'s': 3, 'e': 2, 'terms': terms})
        
    # dW/dz (start:1 -> end:3 via x, y)
    for k in range(c):
        terms = []
        for i in range(a):
            for j in range(b):
                terms.append((Coeffs[i, j, k], [i, a+j])) # x_i, y_j
        base_relations.append({'s': 1, 'e': 3, 'terms': terms})

    # 次数ごとの計算
    dims = []
    
    # Degree 1
    dims.append(num_arrows)
    
    # Degree 2+
    for d in range(2, max_degree + 1):
        # パス生成 (再帰)
        valid_paths = []
        def generate(current_path, current_end):
            if len(current_path) == d:
                valid_paths.append(tuple(current_path))
                return
            for idx, ar in enumerate(arrows):
                if ar['s'] == current_end:
                    generate(current_path + [idx], ar['e'])
        
        for idx, ar in enumerate(arrows):
            generate([idx], ar['e'])
            
        num_paths = len(valid_paths)
        if num_paths == 0:
            dims.append(0)
            continue
            
        # 行列構築
        path_to_col = {p: i for i, p in enumerate(valid_paths)}
        constraints = []
        
        # 関係式の適用 (長さ2の関係式を埋め込む)
        # Left_path + Relation + Right_path = Path of length d
        # Relation length = 2
        remainder = d - 2
        
        for left_len in range(remainder + 1):
            right_len = remainder - left_len
            
            # Left paths
            l_paths = []
            if left_len == 0: l_paths = [([], n) for n in [1,2,3]]
            else:
                temp = []
                def get_l(p, e):
                    if len(p) == left_len: temp.append((p,e)); return
                    for i, ar in enumerate(arrows):
                        if ar['s'] == e: get_l(p+[i], ar['e'])
                for n in [1,2,3]:
                    for i, ar in enumerate(arrows):
                        if ar['s'] == n: get_l([i], ar['e'])
                l_paths = temp
            
            for l_p, l_end in l_paths:
                for rel in base_relations:
                    if rel['s'] == l_end:
                        # Right paths
                        r_paths = []
                        if right_len == 0: r_paths = [[]]
                        else:
                            temp_r = []
                            def get_r(p, e):
                                if len(p) == right_len: temp_r.append(p); return
                                for i, ar in enumerate(arrows):
                                    if ar['s'] == e: get_r(p+[i], ar['e'])
                            for i, ar in enumerate(arrows):
                                if ar['s'] == rel['e']: get_r([i], ar['e'])
                            r_paths = temp_r
                            
                        for r_p in r_paths:
                            row = np.zeros(num_paths)
                            for coeff, term_idxs in rel['terms']:
                                full = tuple(l_p + list(term_idxs) + r_p)
                                if full in path_to_col:
                                    row[path_to_col[full]] = coeff
                            constraints.append(row)
                            
        if not constraints:
            dims.append(num_paths)
        else:
            mat = np.array(constraints)
            # 行列サイズが大きい場合はRank計算が重いが、d=5程度ならOK
            rank = np.linalg.matrix_rank(mat)
            dims.append(max(0, num_paths - rank))
            
        if dims[-1] == 0:
            # 以降はすべて0
            dims.extend([0] * (max_degree - d))
            break
            
    return dims

# --- 実行 ---
# a,b,c を 1から3まで走査
check_algebra_growth_sweep(a_max=3, b_max=3, c_max=3, max_check_degree=5)

# 特定の大きなMarkov Tripleも個別にチェック (例: 2,2,2 や 1,2,5 は範囲外なので)
print("\n--- Spot Check: Specific Markov Triples ---")
def spot_check(a,b,c):
    dims = get_degree_dims(a, b, c, 5)
    print(f"({a},{b},{c}) -> {dims}")

spot_check(2, 2, 2) # k=? (4+4+4 != k*8) -> Not Markov in standard sense but symmetric
spot_check(3, 3, 3) # Markov type
spot_check(1, 2, 5) # Markov (1^2+2^2+5^2 = 30, 3*1*2*5=30 => k=1)
```

```python
import numpy as np

def analyze_markov_tree_deep_fixed(max_degree=5):
    """
    修正版: Markov Tripleの変異をたどり、k=1系列とk=3系列の挙動を比較する
    """
    print(f"--- Deep Analysis of Markov Trees (Checking up to degree {max_degree}) ---")
    print(f"{'Type (k)':<8} | {'Triple (a,b,c)':<20} | {'Sum(1/n)':<8} | {'Status':<10} | {'Dim Sequence'}")
    print("-" * 90)

    def get_dims(a, b, c):
        # 1. Setup Arrows
        arrows = []
        for i in range(a): arrows.append({'s':1, 'e':2})
        for i in range(b): arrows.append({'s':2, 'e':3})
        for i in range(c): arrows.append({'s':3, 'e':1})
        num_arrows = len(arrows)
        
        # 2. Generic Coefficients
        # ランク計算の偶然性を排除するため、ランダム係数を使用
        np.random.seed(a + b + c) 
        Coeffs = np.random.randint(1, 50, size=(a,b,c))
        
        # 3. Relations (Length 2)
        rels = []
        # dW/dx
        for i in range(a):
            t = []
            for j in range(b):
                for k in range(c):
                    t.append((Coeffs[i,j,k], [a+j, a+b+k]))
            rels.append({'s':2, 'e':1, 'terms':t})
        # dW/dy
        for j in range(b):
            t = []
            for i in range(a):
                for k in range(c):
                    t.append((Coeffs[i,j,k], [a+b+k, i]))
            rels.append({'s':3, 'e':2, 'terms':t})
        # dW/dz
        for k in range(c):
            t = []
            for i in range(a):
                for j in range(b):
                    t.append((Coeffs[i,j,k], [i, a+j]))
            rels.append({'s':1, 'e':3, 'terms':t})
            
        # 4. Dimension Calculation
        dims = [num_arrows]
        for d in range(2, max_degree+1):
            # Generate Paths
            valid_paths = []
            def gen(path, end):
                if len(path)==d: valid_paths.append(tuple(path)); return
                for idx, ar in enumerate(arrows):
                    if ar['s'] == end: gen(path+[idx], ar['e'])
            for idx, ar in enumerate(arrows): gen([idx], ar['e'])
            
            n_paths = len(valid_paths)
            if n_paths == 0:
                dims.extend([0]*(max_degree-d+1)); break
            
            # Matrix Construction
            path_map = {p:i for i,p in enumerate(valid_paths)}
            rows = []
            rem = d - 2
            
            for l_len in range(rem+1):
                r_len = rem - l_len
                # Left paths
                l_ps = []
                if l_len==0: l_ps=[([],n) for n in [1,2,3]]
                else:
                    tmp=[]; 
                    def gl(p,e): 
                        if len(p)==l_len: tmp.append((p,e)); return
                        for i,ar in enumerate(arrows): 
                            if ar['s']==e: gl(p+[i], ar['e'])
                    for n in [1,2,3]:
                        for i,ar in enumerate(arrows): 
                            if ar['s']==n: gl([i], ar['e'])
                    l_ps = tmp
                
                for lp, lend in l_ps:
                    for r in rels:
                        if r['s'] == lend:
                            # Right paths
                            r_ps = []
                            if r_len==0: r_ps=[[]]
                            else:
                                tmp=[];
                                def gr(p,e):
                                    if len(p)==r_len: tmp.append(p); return
                                    for i,ar in enumerate(arrows): 
                                        if ar['s']==e: gr(p+[i], ar['e'])
                                for i,ar in enumerate(arrows): 
                                    if ar['s']==r['e']: gr([i], ar['e'])
                                r_ps = tmp
                            
                            for rp in r_ps:
                                row = np.zeros(n_paths)
                                for c_val, t_idx in r['terms']:
                                    full = tuple(lp + list(t_idx) + rp)
                                    if full in path_map: row[path_map[full]] = c_val
                                rows.append(row)
            
            if not rows: dims.append(n_paths)
            else:
                mat = np.array(rows)
                rank = np.linalg.matrix_rank(mat)
                dims.append(max(0, n_paths - rank))
            
            if dims[-1] == 0:
                dims.extend([0]*(max_degree-d)); break
                
        return dims

    # --- Test Cases ---
    
    # k=3 Series (Markov: a^2+b^2+c^2 = 3abc)
    # Mutation: new = 3ab - c
    k3_samples = [
        (1, 1, 1),
        (1, 1, 2),  
        (1, 2, 5),  
        (1, 5, 13)
    ]

    # k=1 Series (Generalized: a^2+b^2+c^2 = 1abc)
    # Mutation: new = 1ab - c
    k1_samples = [
        (3, 3, 3),
        (3, 3, 6),  
        (3, 6, 15), 
        (6, 15, 87)
    ]

    # Run Analysis
    all_samples = [('k=3', t) for t in k3_samples] + [('k=1', t) for t in k1_samples]
    
    for k_type, triple in all_samples:
        a,b,c = triple
        inv_sum = float(1/a + 1/b + 1/c) # floatキャストを追加して修正
        dims = get_dims(a,b,c)
        
        # Classification
        if dims[-1] == 0: status = "Finite"
        elif len(dims)>2 and dims[-1] > dims[-2]: status = "GROWING"
        else: status = "Const/Inf"
        
        print(f"{k_type:<8} | {str(triple):<20} | {inv_sum:.2f}{' ':<4} | {status:<10} | {dims}")

# 実行
analyze_markov_tree_deep_fixed(max_degree=5)
```
